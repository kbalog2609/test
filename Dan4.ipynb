```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan,
                        strategy='median')
imputer.fit(X[['Age','Salary']])
imputer.statistics_
```

```python
#imputer - uči, prepoznaje NaN vrijednosti i zamjeni ih medianom
# vrijednosti koje zelimo da popunis su NaN
# startegija popunjavanja je median
# faza fit(uči) i transform
```

```python
X[['Age','Salary']]. median()
```

```python
X_imputed = imputer.transform(X[['Age','Salary']])
X_imputed
```

```python
X_imp= pd.DataFrame(X_imputed)
X_imp
```

```python
X_imp = pd.DataFrame (X_imputed,
                      columns = ['Age','Slarly'],
                      index=X.index)
X_imp
```
