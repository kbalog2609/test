## Kreiramo novi stupac

```python
df['WminY'] = df['W'] - df['Y']
```

```python
df
```

```python
df['logX'] = np.log10(df['X'])
df
```

```python
def ph(c):
  return -np.log10(c)
```

```python
# apply

df['X'].apply(ph)
```

```python
# lambda - anonimna (bezimena) funkcija
# lambda argument(i): izraz

df['Y'].apply(lambda x: -np.log10(x))
```

```python
def above_LOD(x):

  if x > 0.01:
    return 'iznad LOD'
  else:
    return 'XXX'
```

```python
# applymap

df.applymap(above_LOD)
```

```python
# 1

df.applymap(lambda x: 2*x + 100)
```

```python
# 2

def fun(x):
  return 2*x+100

df.applymap(fun)
```

```python
df['aboveLOD'] = df['X'].apply(lambda x: x if x > 0.01 else 0.0001)
df
```

```python
df = pd.DataFrame(data=randn(5,4),
                  index=['A','B','C','D','E'],
                  columns=['W','X','Y','Z'])
df
```

```python
lod = 0.001

df[df > lod].count(axis=1)
```

```python
# logicki operatori
# & (AND)
# | (OR) alt gr + W
# ~ (NOT) alt gr + 1

df[(df['W'] > 0) | (df['Y'] > 0)]
```
